```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

btc = yf.download('BTC-USD', period='5d', interval='1d')
eth = yf.download('ETH-USD', period='5d', interval='1d')

btc['Crypto'] = 'Bitcoin'
eth['Crypto'] = 'Ethereum'

crypto_df = pd.concat([btc, eth])
crypto_df = crypto_df[['Crypto', 'Close']].reset_index()

for crypto in crypto_df['Crypto'].unique():
    subset = crypto_df[crypto_df['Crypto'] == crypto]
    plt.plot(subset['Date'], subset['Close'], label=crypto)

plt.title("Crypto Prices (Last 5 Days)")
plt.xlabel("Date")
plt.ylabel("Price in USD")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
